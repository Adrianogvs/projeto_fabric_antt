# 00 · Config — Logging

> **Notebook compartilhado — não executar diretamente.**  
> Chamado via `%run 00_nb_config` pelos outros notebooks.

**Variáveis exigidas no namespace do chamador antes do `%run`:**

| Variável | Tipo | Exemplo |
|---|---|---|
| `NOTEBOOK_NAME` | `str` | `"bronze_to_silver"` |
| `LOG_LEVEL` | `str` | `"INFO"` |

```python
import logging

logging.basicConfig(
    level=getattr(logging, LOG_LEVEL),
    format="%(asctime)s | %(levelname)-8s | %(message)s",
    datefmt="%H:%M:%S",
    force=True,
)
log = logging.getLogger(NOTEBOOK_NAME)
log.info("Logging configurado | notebook=%s | level=%s", NOTEBOOK_NAME, LOG_LEVEL)
```
